```python
# ======== 1. تحميل المكتبات الضرورية ========
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# ======== 2. قراءة بيانات التدريب ========
train_image_dir = 'a. Training Set'
train_labels_file = 'a. IDRiD_Disease Grading_Training Labels.csv'

# تحميل بيانات التدريب
train_df = pd.read_csv(train_labels_file)

# حذف الأعمدة الغير مهمة
unnamed_cols = [col for col in train_df.columns if col.startswith('Unnamed')]
train_df = train_df.drop(columns=unnamed_cols)
train_df = train_df.drop(columns=['Risk of macular edema '])

# إضافة عمود التصنيف الثنائي
train_df['Label'] = train_df['Retinopathy grade'].apply(lambda x: 0 if x == 0 else 1)

# إضافة امتداد .jpg لكل اسم صورة
train_df['Image name'] = train_df['Image name'].astype(str) + '.jpg'

# ======== 3. تجهيز بيانات التدريب ========
img_width, img_height = 224, 224

X_train = []
y_train = []

for index, row in train_df.iterrows():
    img_path = os.path.join(train_image_dir, row['Image name'])
    if os.path.exists(img_path):
        img = load_img(img_path, target_size=(img_width, img_height))
        img_array = img_to_array(img)
        img_array = preprocess_input(img_array)  # تجهيز للصورة حسب ResNet50
        X_train.append(img_array)
        y_train.append(row['Label'])

X_train = np.array(X_train)
y_train = np.array(y_train)

print(f'عدد صور التدريب الجاهزة: {X_train.shape[0]}')

# ======== 4. قراءة بيانات الاختبار ========
test_image_dir = 'b. Testing Set'
test_labels_file = 'b. IDRiD_Disease Grading_Testing Labels.csv'

# تحميل بيانات الاختبار
test_df = pd.read_csv(test_labels_file)

# إضافة امتداد .jpg لكل اسم صورة
test_df['Image name'] = test_df['Image name'].astype(str) + '.jpg'

# إضافة عمود التصنيف الثنائي
test_df['Label'] = test_df['Retinopathy grade'].apply(lambda x: 0 if x == 0 else 1)

# ======== 5. تجهيز بيانات الاختبار ========
X_test = []
y_test = []

for index, row in test_df.iterrows():
    img_path = os.path.join(test_image_dir, row['Image name'])
    if os.path.exists(img_path):
        img = load_img(img_path, target_size=(img_width, img_height))
        img_array = img_to_array(img)
        img_array = preprocess_input(img_array)
        X_test.append(img_array)
        y_test.append(row['Label'])

X_test = np.array(X_test)
y_test = np.array(y_test)

print(f'عدد صور الاختبار الجاهزة: {X_test.shape[0]}')

# ======== 6. بناء النموذج (Transfer Learning - ResNet50) ========
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))

# تثبيت أوزان ResNet50 (عدم تدريبها من جديد)
for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

model.summary()

# ======== 7. تدريب النموذج ========
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),  # بيانات الاختبار كـ validation
    epochs=20,
    batch_size=32,
    callbacks=[early_stop]
)

# ======== 8. رسم منحنى الدقة والخسارة ========
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid()
plt.show()

# ======== 9. رسم Confusion Matrix ========
# توقع النتائج
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype('int32')

# رسم Confusion Matrix بشكل صحيح
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
